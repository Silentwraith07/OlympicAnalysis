```
def is_prime(num):
  for i in range(2,num):
    if num % i == 0:
      print(' not a Prime Number')
      break
    else:
      print('Prime Number')
      break
```

```
is_prime(7)
```

```
def even_comp(arg1,arg2):
  if(arg1%2 == 0 and arg2%2 == 0):
    if arg1 > arg2:
      print(arg2 , 'is smaller')
    elif arg1< arg2:
      print(arg1 , 'is smaller')
    else:
      print('both are equal')
  else:
    print('enter even numbers')
```

```
even_comp(1029384,2)
```

```
def divisible(num):
  return (num / 7 and num / 7) == 1
```

```
#create a function in which you will find values divisible by 7 & 9 only  then create a list of 500:1000 use the filter command to find the true values only
```

```
lstv = []
for num in range(500,1000):
  lstv.append(num)
  
list(filter(divisible,lstv))
```
